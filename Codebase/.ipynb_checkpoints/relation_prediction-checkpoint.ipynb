```python
import os
import warnings
import random
import pandas as pd
from itertools import combinations
import sys
import networkx as nx
import spacy
import traceback
import pickle
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder

warnings.filterwarnings('ignore')

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)

random.seed(100)
```

```python
dataset = "kbp37"
```

```python
features = pd.read_csv("/Users/anishajauhari/Desktop/Sem 4/Independent Study /ResponsibleRelationExtraction/Features/full_features_"+dataset+".csv", index_col=0)
features.head(10)
```

```python
features["rel subtype"].unique()
```

```python
column_names = ['sent_index', 'sentence_length', 'root_word', 'root_index',
       'entity_1_index', 'entity_2_index', 'entity_distance',
       'no_words_before_entity_1', 'no_words_after_entity_2',
       'entity_1_root_distance', 'entity_2_root_distance', 'entity 1 name',
       'entity 2 name', 'entity_type_1', 'entity_type_2', 'entity_pos_1',
       'entity_pos_2', 'entity_dep_1', 'entity_dep_2', 'entity_tag_1',
       'entity_tag_2', 'shortest_distance', 'rel type']
features = features[column_names]
features.rename(columns={"rel type": "label"}, inplace=True)
```

```python
# features_with_labels = pd.read_csv("features_"+dataset+".csv")
X = features.iloc[:,:-1]
y_original =  features.iloc[:,-1]
X = X.fillna("")
print(y_original)
onehot_encoder = OneHotEncoder(sparse=False)
X = onehot_encoder.fit_transform(X)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y_original)
print(y)

x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
unique, count = np.unique(y_original, return_counts = True)
dict(zip(unique, count))
```

```python
def logisticMultiClass(x_train, x_test, y_train, y_test):
    model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)    
    pickle.dump(model, open("et_logistic_model_"+dataset+".sav", "wb"))
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
    print("Precision: ", metrics.precision_score(y_test, y_pred, labels=np.unique(y), average="weighted"))
    print("Recall: ", metrics.recall_score(y_test, y_pred, labels=np.unique(y), average="weighted"))
    print("F-Score: ", metrics.f1_score(y_test, y_pred, labels=np.unique(y), average="weighted"))
    print(classification_report(y_test, y_pred))
```

```python
logisticMultiClass(x_train, x_test, y_train, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
def random_forest(x_train, x_test, y_train, y_test):
    print("--------Random Forest-----------")
    clf = RandomForestClassifier(n_estimators=10, class_weight='balanced', n_jobs=10)
    clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
#     print("Features: ", columns[:-1])
    print("Coefficients: ", clf.feature_importances_)
    pickle.dump(clf, open("et_rf_model_"+dataset+".sav", "wb"))

    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
    print("Precision: ", metrics.precision_score(y_test, y_pred, labels=np.unique(y), average="weighted"))
    print("Recall: ", metrics.recall_score(y_test, y_pred, labels=np.unique(y), average="weighted"))
    print("F-Score: ", metrics.f1_score(y_test, y_pred, labels=np.unique(y), average="weighted"))
    print(classification_report(y_test, y_pred))
```

```python
random_forest(x_train, x_test, y_train, y_test)
```

```python
import xgboost as xgb
from sklearn.metrics import auc, accuracy_score, confusion_matrix, mean_squared_error

def xgboost(x_train, x_test, y_train, y_test):
    xgb_model = xgb.XGBClassifier(objective="multi:softmax", random_state=42)
    xgb_model.fit(x_train, y_train)
    y_pred = xgb_model.predict(x_test)
    print(y_pred)
    pickle.dump(xgb_model, open("et_xgboost_"+"_model_"+dataset+".sav", "wb"))
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
    print("Precision: ", metrics.precision_score(y_test, y_pred, labels=np.unique(y), average="weighted"))
    print("Recall: ", metrics.recall_score(y_test, y_pred, labels=np.unique(y), average="weighted"))
    print("F-Score: ", metrics.f1_score(y_test, y_pred, labels=np.unique(y), average="weighted"))
    print(classification_report(y_test, y_pred))
```

```python
xgboost(x_train, x_test, y_train, y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier

def adaboost(x_train, x_test, y_train, y_test):
    ada = AdaBoostClassifier(n_estimators=100, random_state=0)
    ada.fit(x_train, y_train)
    y_pred = ada.predict(x_test)
    pickle.dump(ada, open("et_adaboost_"+"_model_"+dataset+".sav", "wb"))
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
    print("Precision: ", metrics.precision_score(y_test, y_pred, labels=np.unique(y), average="weighted"))
    print("Recall: ", metrics.recall_score(y_test, y_pred, labels=np.unique(y), average="weighted"))
    print("F-Score: ", metrics.f1_score(y_test, y_pred, labels=np.unique(y), average="weighted"))
    print(classification_report(y_test, y_pred))
    
```

```python
adaboost(x_train, x_test, y_train, y_test)
```
